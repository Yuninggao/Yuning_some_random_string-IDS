```python
# import the  packages
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
# read dermatology csv file and print the file

df = pd.read_csv("dermatology_csv.csv")

df
```

```python
# summarize the numeric attributes

df.describe()
```

```python
# check any missing values

df.isnull().sum()


```

```python
# age column has 8 missing value,so I use mean replace missing value

df['age'] = df['age'].fillna(df['age'].mean())

df.isnull().sum()



```

```python
# change all class name as disease and plotting boxplot, x as disease and y as age

# the group of between about age 28 to 52  mainly has the first  disease 
# the group of between about age 25 to 45  mainly has the second disease 
# the group of between about age 40 to 51  mainly has the thrid  disease
# the group of between about age 28 to 45  mainly has the fourth disease 
# the group of between about age 29 to 48  mainly has the fifth disease 
# the group of between about age 10 to 13  mainly has the sixth disease 

df.rename(columns = {'class':'disease'})
df

sns.boxplot(x = df['disease'], 
            y = df['age'])




```

```python
# sorting the values using descening trend
corr_matrix = df.corr()
corr_matrix['disease'].sort_values(ascending = False)

```

```python
# create a contigency table using top 5  feature with disease

top_disease =['fibrosis_of_the_papillary_dermis','follicular_papules','perifollicular_parakeratosis','follicular_horn_plug','exocytosis'  ]

for i in top_disease:
    crosstable = pd.crosstab(index = [df[i]],columns=df['disease'], margins= True)
    
print(crosstable)
```

```python
# create a new column as diseas 1 to exact diseas equal to 1


df["disease_1"] = np.where(df["disease"] == 1,1,0)
df

```

```python
# create a linear regression model using disease_1 and age

from sklearn.linear_model import LinearRegression

# save all the rows of column disease_1 as x value
X = df.loc[:, 'disease_1'].values.reshape(-1,1)

# save all the rows of column age as y value
Y = df.loc[:, 'age'].values.reshape(-1,1)

# Create the model
LRmodel = LinearRegression()

# Perform linear regression
LRmodel.fit(X, Y) 

# Make predictions
Y_pred = LRmodel.predict(X)

# Visualize the dataset and the regression line:
plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
plt.show()
```

```python
# The equation of the line is: 

print("The slope: ", LRmodel.coef_)
print("The intercept: ", LRmodel.intercept_)
```

```python
# Error analysis

from sklearn.metrics import mean_squared_error, r2_score

print("MSE: ", mean_squared_error(Y, Y_pred))
print("R2: ", r2_score(Y, Y_pred))
```

```python
# create  c5.0 classifier
# The C5.0 tree; try different values for max_depth and max_leaf_nodes
c5_classifier = DecisionTreeClassifier(criterion = "entropy", 
                                       max_leaf_nodes=5,
                                       random_state = 120).fit(X_train,y_train)


# Control the figure size
from matplotlib.pyplot import figure
figure(figsize=(12, 10), dpi=80)


# Plot the tree

tree.plot_tree(c5_classifier, feature_names=X.columns.values.tolist(), 
                               
                                filled=True)


# Print the text version of the tree
text_representation = tree.export_text(c5_classifier, 
                                       feature_names=X.columns.values.tolist())
print(text_representation)
```
