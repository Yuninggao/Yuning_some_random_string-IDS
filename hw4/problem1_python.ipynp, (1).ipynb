```python
# read the csv file

import pandas as pd


df=pd.read_csv('adult_training.csv')
df
```

```python
# cross tabulation of workclass and sex
ctab = pd.crosstab( df.workclass, df.sex, margins= True)

ctab

```

```python
# create the barplot
barplot = ctab.plot.bar(rot=50)
```

```python
#cross tabulation of sex and marital
ctab = pd.crosstab( df.sex, df.marital_status, margins= True)

ctab
```

```python
#create the barplot
barplot = ctab.plot.bar(rot=50)
```

```python
# create another dataframe and age is over 40
df2 = df[(df.age > 40)]

df2
```

```python
# cross tabulation of sex  and marital that age is over 40
ctab = pd.crosstab( df2.sex, df2.marital_status, margins= True)

ctab
```
