```r
# Let's read in the data file

library(readr)

df <- read_csv("nutrition_subset.csv")

df
```

```r
# find top 5 items highest saturated-fat 
head(df[order(df$saturated_fat, decreasing=TRUE),1:3], 5)
```

```r
# create a new variable  saturated_ fat_ per_gram,
 saturated_fat_per_gram <- df$saturated_fat/df$weight_in_grams

#add column called  saturated_ fat_ per_gram 
df$saturated_fat_per_gram  <-  saturated_fat_per_gram 


df 
```

```r
# find top 5 items highest saturated_fat_per_gram 
head(df[order(df$saturated_fat_per_gram,decreasing=TRUE),1:5],5)
```

```r
# using scale function to calculate the z-score
df$scaled_saturated_fat_per_gram <- scale(x = df$saturated_fat_per_gram)

summary(df$scaled_saturated_fat_per_gram)

```

```r
# find the high outliers in saturated_fat_per_gram .

three_sigma_rule <- (df$scaled_saturated_fat_per_gram > 3 )

cat ("There are", sum(three_sigma_rule), "high outliers in saturated_fat_per_gram \n\n")
```

```r
# find the low outliers in saturated_fat_per_gram .

three_sigma_rule <- (df$scaled_saturated_fat_per_gram < -3 )

cat ("There are", sum(three_sigma_rule), "low outliers in saturated_fat_per_gram \n\n")
```

```r
#  create density Plot
df_Density <- density(df$saturated_fat_per_gram) 
plot(df_Density) 
```

```r
# using scale function to calculate the z-score 
df$weight_in_grams <- scale(x = df$weight_in_grams)

summary(df$weight_in_grams)
```

```r
# find the outlier of weight_in_grams
three_sigma_rule <- (df$weight_in_grams > 3 | df$weight_in_grams< -3)

cat ("There are", sum(three_sigma_rule), "outliers in weight_in_grams \n\n")
```

```r
# using scale function to calculate the z-score 
df$saturated_fat <- scale(x = df$saturated_fat)

summary(df$saturated_fat)
```

```r
# find the outlier of saturated_fat
three_sigma_rule <- (df$saturated_fat > 3 | df$saturated_fat< -3)

cat ("There are", sum(three_sigma_rule), "outliers in saturated_fat \n\n")
```

```r
# using scale function to calculate the z-score 
df$cholesterol <- scale(x = df$cholesterol)

summary(df$cholesterol)
```

```r
# find the outlier of cholesterol
three_sigma_rule <- (df$cholesterol > 3 | df$cholesterol< -3)

cat ("There are", sum(three_sigma_rule), "outliers in cholesterol \n\n")
```
