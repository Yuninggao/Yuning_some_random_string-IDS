```r
# read the file using r
library(readr)

df <- read_csv("adult_training.csv")

df
```

```r
# cross tabulation of workclass and sex
r_crosstab <- table(df$workclass,df$sex)
ftable(r_crosstab)
```

```r
# create a barplot

colors <- c("navy","orange","brown","blue","yellow","red","green","pink")
barplot(r_crosstab, col = colors, beside = TRUE)
legend("topleft",legend = c ("Federal-gov","Local-gov","Never-worked","Private","Self-emp-inc","self-emp-not-inc","State-gov","Without-pay"),fill = colors)
```

```r
# cross tabulation of sex and marital status
r_crosstab <- table(df$sex,df$marital_status)
ftable(r_crosstab)
```

```r

# create a barplot with sex and marrital status

colors <- c("navy","orange")
barplot(r_crosstab, col = colors, beside = TRUE)
legend("topright",legend = c ("Female","Male"),fill = colors)
```

```r
# create a new dataframe after filtering age over 40

library(dplyr)

df2 <- df %>% filter(age > 40)

df2
```

```r
r_crosstab <- table(df2$sex,df2$marital_status)
ftable(r_crosstab)
```
