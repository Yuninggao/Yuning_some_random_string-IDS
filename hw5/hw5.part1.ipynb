```r
# read the file

library(readr)

df <- read_csv('https://raw.githubusercontent.com/bforoura/AML/master/Company_data.csv')

df
```

```r
# read first 10 datas

head(df, 10)
```

```r
# describe all the dataframe
summary(df)
```

```r
# compute the standard correlation coefficient using Pearson method between 
#,sales and sales, tv and sales, radio and sales, newspaper and sales

cor.test(df$Sales,df$Sales,method=c("pearson"))
cor.test(df$TV,df$Sales,method=c("pearson"))
cor.test(df$Radio,df$Sales,method=c("pearson"))
cor.test(df$Newspaper,df$Sales,method=c("pearson"))

```

```r
# get intercept and slope value using linear model regression

model <- lm(df$Sales~df$TV)
model
```

```r
# get mse value and  r^2 value
mean((df$Sales- predict(model))^2)
summary(model)$r.squared
```
